# Data Cleaning

```python
holdout.drop(columns =['Unnamed: 0'], inplace = True)
```

```python
for col in holdout.columns:  # Converting all cols from type str to int
    holdout[col] = holdout[col].astype(int)
```

```python
holdout.rename(columns = {'X1': 'credit_given', 'X2': 'gender', 'X3': 'education', 
                    'X4':'marital_status', 'X5':'age', 'X6':'repay_status_sep', 
                    'X7':'repay_status_aug', 'X8':'repay_status_jul', 
                    'X9':'repay_status_jun', 'X10':'repay_status_may', 
                    'X11':'repay_status_apr', 'X12':'bill_amt_sep', 
                    'X13':'bill_amt_aug', 'X14':'bill_amt_jul', 
                    'X15':'bill_amt_jun', 'X16':'bill_amt_may', 
                    'X17':'bill_amt_apr', 'X18':'paid_amt_sep',
                    'X19':'paid_amt_aug', 'X20':'paid_amt_jul',
                    'X21':'paid_amt_jun', 'X22':'paid_amt_may', 
                    'X23':'paid_amt_apr', 'Y':'default_status'}, inplace = True)
```

```python
holdout['education'] = holdout['education'].apply(lambda x: 4 if x not in [1,2,3,4] else x)
```

```python
holdout['marital_status'] = holdout['marital_status'].apply(lambda x: 3 if x not in [1,2,3] else x)
```

```python
holdout['gen_edu'] = holdout['gender'] * holdout['education']
```
